```python
# Linear Regression Example in Python

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate synthetic data
np.random.seed(42)  # reproducibility
x = np.linspace(0, 10, 30)                       # independent variable
y = 3.5 * x + 7 + np.random.normal(0, 5, size=30) # dependent variable with noise

# 2. Perform linear regression using scipy
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# 3. Create fitted values for plotting
y_pred = slope * x + intercept

# 4. Plot the data points and regression line
plt.figure(figsize=(8,5))
plt.scatter(x, y, label="Observed data", color="blue")
plt.plot(x, y_pred, label=f"Fitted line: y = {slope:.2f}x + {intercept:.2f}", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Example")
plt.legend()
plt.grid(True)
plt.show()

# 5. Print regression attributes
print(f"Intercept (α): {intercept:.4f}")
print(f"Slope (β): {slope:.4f}")
print(f"Pearson correlation coefficient (r): {r_value:.4f}")
print(f"R-squared (r^2): {r_value**2:.4f}")
print(f"p-value for slope: {p_value:.4e}")
print(f"Standard error of slope: {std_err:.4f}")

```
