```python

import numpy as np
import matplotlib as plt
import pandas as pd
import yfinance as yf
```

```python


# Download NQ futures data for past year
nq = yf.download("NQ=F", period="1y", interval="1d")

# Compute daily percent returns
nq["Daily Return %", ] = nq.pct_change() * 100

# Drop NA values
nq = nq.dropna()

# Plot daily returns
plt.figure(figsize=(12,6))
plt.plot(nq.index, nq["Daily Return %"], label="Daily Return %")
plt.title("NQ Futures - Daily Returns (Past 1 Year)")
plt.xlabel("Date")
plt.ylabel("Return (%)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import erf, sqrt, exp, pi

# Parameters
mu = 0
sigma = 1

# Define the normal PDF
def normal_pdf(x, mu=0, sigma=1):
    return (1 / (sigma * sqrt(2 * pi))) * exp(-((x - mu)**2) / (2 * sigma**2))

# Define the normal CDF using the error function erf()
def normal_cdf(x, mu=0, sigma=1):
    return 0.5 * (1 + erf((x - mu) / (sigma * sqrt(2))))

# Generate x values and compute CDF
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
cdf_values = [normal_cdf(val, mu, sigma) for val in x]

# Plot
plt.plot(x, cdf_values, label=f'Normal CDF (μ={mu}, σ={sigma})')
plt.title("Cumulative Distribution Function of Normal Distribution")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)
plt.legend()
plt.show()
```
