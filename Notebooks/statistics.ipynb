```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
mu = 4  # mean (λ) of the Poisson distribution
x = np.arange(0, 15)  # range of x values

# PMF values
pmf = poisson.pmf(x, mu)

# Plot
plt.stem(x, pmf, basefmt=" ")
plt.title(f'Poisson Distribution (μ = {mu})')
plt.xlabel('Number of events (k)')
plt.ylabel('Probability P(X = k)')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import lognorm

sigma = 0.9  # Shape parameter (standard deviation of the underlying normal)
scale = np.exp(0)  # exp(mean of the underlying normal), here mean = 0

# X range
x = np.linspace(0.001 , 10, 1000)

# PDF
pdf = lognorm.pdf(x, s=sigma, scale=scale)

# Plot
plt.plot(x, pdf, label=f'Lognormal(μ=0, σ={sigma})', color='darkorange')
plt.title('Lognormal Distribution (PDF)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```
